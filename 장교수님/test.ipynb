```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y_tr = y[:4500]
y_ts = y[4500:]
x_tr = x[:4500]
x_ts = x[4500:]
```

```python
rf.score(x_ts, y_ts)
```
